```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
```

```python
# Source Image Size
img = cv2.imread('BikeShot1.jpg')
dimensions = img.shape
 
    # height, width, number of channels in image

height = img.shape[0]
width = img.shape[1]
channels = img.shape[2]
print(height, width, channels)
```

```python
# Resizing of Image
# dsize
src=img
dsize = (425, 425)

# resize image
output = cv2.resize(src, dsize, interpolation = cv2.INTER_AREA)
height = output.shape[0]
width = output.shape[1]
print(height, width)
```

```python
plt.imshow(output)
```

```python
norm_output = np.zeros((425, 425))
norm_output = cv2.normalize(output,np.zeros((output.shape[0],output.shape[1],output.shape[2])),0.485,0.229,dtype=cv2.CV_32F,norm_type=cv2.NORM_MINMAX)
```

```python
plt.imshow(norm_output)
```
